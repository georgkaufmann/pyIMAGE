# Image processing: Creating functions

In this notebook, we create 2D functions and save them in the different channels
of a color image.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import PIL
```

----
## 1D functions

As basis, we define two 1D fucntions:
$$
\begin{array}{rcl}
f_1(x) &=& a\sin(x) \\
f_2(x) &=& a e^{bx}
\end{array}
$$    

```python
def f1(x,a=1):
    f = a*np.sin(x)
    return f

def f2(x,a=1,b=0.1):
    f = a*np.exp(b*x)
    return f
```

```python
x = np.linspace(-2*np.pi,2*np.pi,51)

plt.plot(x,f1(x),label='$f_1(x)=a \sin(x)$')
plt.plot(x,f2(x),label='$f_2(x)=a \exp(bx)$')
plt.grid()
plt.legend()
```

----
## 2D function

We use the two 1D function, one in $x$- one in $y$-direction, to create two 2D functions,
plus an additional 2D function.
$$
\begin{array}{rcl}
F_1(x,y) &=& a\sin(x) \\
F_2(x,y) &=& a e^{by} \\
F_3(x,y) &=& \left( 1 - \frac{x}{2} + x^5 + y^3 \right) e^{(-x^2 - y^2)}
\end{array}
$$  

```python
def F1(x,y,a=1):
    f = a*np.sin(x)
    return f

def F2(x,y,a=1,b=0.1):
    f = a*np.exp(b*y)
    return f

def F3(x,y):
    # two-dimensional function
    f3 = (1. - x/2. + x**5 + y**3) * np.exp(-x**2 - y**2)
    return f3
```

```python
nx=51; ny=21
x = np.linspace(-2*np.pi,2*np.pi,nx)
y = np.linspace(-2,2,ny)
X,Y = np.meshgrid(x,y)

fig,axs = plt.subplots(3,1,figsize=(10,8))
axs[0].set_title('$F_1(x,y)=a \sin(x)$')
axs[0].set_ylabel('y')
cbar0=axs[0].contourf(X,Y,F1(X,Y))
plt.colorbar(cbar0,location='right',shrink=0.9)

axs[1].set_title('$F_2(x,y)=a \exp(bx)$')
axs[1].set_ylabel('y')
cbar1=axs[1].contourf(X,Y,F2(X,Y))
plt.colorbar(cbar1,location='right',shrink=0.9)

axs[2].set_title('$F_3(x,y) = ( 1 - \\frac{x}{2} + x^5 + y^3 ) e^{(-x^2 - y^2)}$')
axs[2].set_xlabel('x')
axs[2].set_ylabel('y')
cbar2=axs[2].contourf(X,Y,F3(X,Y))
plt.colorbar(cbar2,location='right',shrink=0.9)

plt.tight_layout()
```

These functions are calculated for $x \in [-2\pi,2\pi]$ and $y \in [-2,2]$, and then 
normalized to fit into the [0,255] domain, the 8-bit representation for color images.

```python
# get min/max values
minx1 = F1(X,Y).min()
maxx1 = F1(X,Y).max()
minx2 = F2(X,Y).min()
maxx2 = F2(X,Y).max()
minx3 = F3(X,Y).min()
maxx3 = F3(X,Y).max()

# normalize functions to [0,255]
r=255*(F1(X,Y) - minx1)/(maxx1-minx1) 
g=255*(F2(X,Y) - minx2)/(maxx2-minx2) 
b=255*(F3(X,Y) - minx3)/(maxx3-minx3)

r = r.astype('int')
g = g.astype('int')
b = b.astype('int')

print (r.min(),r.max())
print (g.min(),g.max())
print (b.min(),b.max())
print(r.shape)
```

----
## Save to image channels

Create a 3D array `image_arr` with the first two dimensions of the 2D functions and 
the third dimension for the channels (rgba).

```python
image_arr = np.zeros([ny,nx,4],dtype=np.uint8)
image_arr[:,:,0] = np.flip(r,axis=0)
image_arr[:,:,1] = np.flip(g,axis=0)
image_arr[:,:,2] = np.flip(b,axis=0)
image_arr[:,:,3] = 255

print (type(image_arr),image_arr.dtype,image_arr.ndim,image_arr.shape)
print(image_arr[0][0])
print(image_arr[20][0])
```

Convert the numpy array to an imagem show and save it to disk.

```python
result = PIL.Image.fromarray((image_arr).astype(np.uint8))
print(result)
# show image
fig,axs = plt.subplots(1,1,figsize=(10,5))
axs.set_title('Original ('+str(image_arr.shape[0])+'x'+str(image_arr.shape[1])+')')
axs.imshow(result)

result.save('Functions2D.png')
```
