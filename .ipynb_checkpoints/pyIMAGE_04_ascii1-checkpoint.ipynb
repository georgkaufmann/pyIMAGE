# Image processing: Ascii art

```python
import matplotlib.pyplot as plt
import numpy as np
import PIL
import PIL.ImageDraw
import PIL.ImageFont
import PIL.ImageChops
```

We first define an image, then check with the external `file` command its content:
- PNG ([portable network graphics](https://de.wikipedia.org/wiki/Portable_Network_Graphics)) data format
- 4928 x 3264 pixel
- 8-bit depth
- 4 channels, red/green/(blue and an alpha channel (RGBA) 

###  `Image.read`

After reading the file with the `Image.read` function from PIL, we receive a
PIL object, whose features can be analysed. e.g. with `.format`, `.size`, or `.mode`.

For using our own image manipulation within numpy and scipy, we need to convert
the PIL object first to a numpy array, using `asarray`.


Then we obtain a 3D numpy array, with **integer** numbers, the first and second dimension the 
pixel rows and columns, and the third dimension the color (rgb) and tansparency
(a) channels.

**Note:** Now, the the png file format
is integer within the range [0,255].

```python
infile='images/Amrum.png'
#infile='images/Redwingblackbird.jpg'
image = PIL.Image.open(infile)
print (image.format,image.size,image.mode)
plt.imshow(image)
```

```python
image = image.resize((200,100))
print (image.format,image.size,image.mode)
plt.imshow(image)
```

```python
# convert to ndarray
image_color = np.asarray(image)
print (type(image_color),image_color.dtype,image_color.ndim,image_color.shape)
print(image_color[0][0])
```

----
## Grayscale image

Define an ascii table with ten characters ...

```python
ascii_table = "@JD%*P+Y$,."
ascii_table_len = len(ascii_table)
print(type(ascii_table))
print(ascii_table_len,ascii_table)
```

Convert color image to grayscale image, then downscale the [0,255] range with `ascii_table_len`.

```python
image_gray = np.zeros([image_color.shape[0],image_color.shape[1]])
image_gray[:,:] = ((0.2126*image_color[:,:,0]
                    +0.7152*image_color[:,:,1]
                    +0.0722*image_color[:,:,2])/255*ascii_table_len)
image_gray = image_gray.astype('uint8')
print(image_color.shape)
print(image_gray.shape,image_gray.min(),image_gray.max())
```

```python
image_gray_ascii = np.zeros([image_gray.shape[0],image_gray.shape[1]]).astype('str')
for i in range(image_gray.shape[0]):
    for j in range(image_gray.shape[1]):
        #print(image_gray[i,j],ascii_table[image_gray[i,j]])
        image_gray_ascii[i,j] = str(ascii_table[image_gray[i,j]])
```

```python
outfile = open('Amrum1.ascii','w')
for i in range(image_gray_ascii.shape[0]):
    line=""
    for j in range(image_gray_ascii.shape[1]):
        line += str(image_gray_ascii[i,j])
    print(line,file=outfile)
    #print(line)
outfile.close()
```

Convert text lines to image lines, using the `ImageDraw.Draw` function from `PIL`.
**Note:** Use a **monospace** font!

Print to image:
- open empty image file, dimensions in vertical direction from line offset
- define font as a monospaced font
- write ascii lines as text into new image
- crop image with the `trim` function

[Check for fonts here ...](https://fontesk.com/monocode-typeface)

```python
def trim(im):
    """
    trim an image by determining the bounding box and then crop it
    """
    bg = PIL.Image.new(im.mode, im.size, im.getpixel((0,0)))
    diff = PIL.ImageChops.difference(im, bg)
    diff = PIL.ImageChops.add(diff, diff, 2.0, -100)
    bbox = diff.getbbox()
    if bbox:
        return im.crop(bbox)
```

```python
width=20*image_gray_ascii.shape[1]
height=20*image_gray_ascii.shape[0]
img = PIL.Image.new('RGB', (width,height),color=(255,255,255))
font = PIL.ImageFont.truetype("monocode/ttf/monocode-bold.ttf",18)
d = PIL.ImageDraw.Draw(img)
for i in range(image_gray_ascii.shape[0]):
    line=""
    for j in range(image_gray_ascii.shape[1]):
        line += str(image_gray_ascii[i,j])
    d.text((0,20*i), line, fill=(0, 0, 0),font=font)
    
img_trimmed = trim(img)
plt.imshow(img_trimmed)
img_trimmed.save('Amrum1_ascii.jpg')
```
