# Image processing: Geophysical data -1-

In this notebook, we read in several irregular geophysical data sets, grid them,
scale them, and finally save them in the different channel of an image.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import scipy.interpolate
import PIL
```

----
## Reading irregular data

The data sets are given as easting, northing, elevation, and field value data for
- Bouguer gravity [mGal]
- Total field [nT]
- Self-potential [mV]
- Topography [m]

These data sets are read with the function `loadtxt` from the `numpy` package into 2D numpy arrays.

```python
!ls *WALPENREUTH2016.utm; head GRAVITY_WALPENREUTH2016.utm
```

```python
# read geophysical data
file_grav='GRAVITY_WALPENREUTH2016.utm'
data_grav=np.loadtxt(fname=file_grav, dtype='float',skiprows=0)
print('data_grav: ',type(data_grav),data_grav.shape)
file_mag='MAGNETICS_WALPENREUTH2016.utm'
data_mag=np.loadtxt(fname=file_mag, dtype='float',skiprows=0)
print('data_mag: ',type(data_mag),data_mag.shape)
file_sp='SP_WALPENREUTH2016.utm'
data_sp=np.loadtxt(fname=file_sp, dtype='float',skiprows=0)
print('data_sp: ',type(data_sp),data_sp.shape)
```

----
## Griding data to regular grid
From all data sets, the min/max values for UTM coordinates are used to find the rectangle,
where all data sets fit it.

```python
# min/max coordinates
eastmin = np.min([data_grav[:,0].min(),data_mag[:,0].min(),data_sp[:,0].min()])
eastmax = np.max([data_grav[:,0].max(),data_mag[:,0].max(),data_sp[:,0].max()])
northmin = np.min([data_grav[:,1].min(),data_mag[:,1].min(),data_sp[:,1].min()])
northmax = np.max([data_grav[:,1].max(),data_mag[:,1].max(),data_sp[:,1].max()])

print('min/max easting:  ',eastmin,eastmax)
print('min/max northing: ',northmin,northmax)
```

A **regular** 2D grid is set up for the rectangular area, first defining 1D arrays for the
easting and northing coordinates with `np.linspace`, then extending the 1D arrays two 2D arrays,
using `meshgrid` from the `matplotlib.pyplot` library.

```python
# set up 2D grid
nx=100; ny=50
nx=200; ny=100
x,dx = np.linspace(eastmin,eastmax,nx,retstep=True)
y,dy = np.linspace(northmin,northmax,ny,retstep=True)
X,Y = np.meshgrid(x,y)
print(X.shape,Y.shape,dx,dy)
```

The **irregular field data** are gridded onto the **regular grid**, using the `griddata` function
from the `scipy.interpolate` package.

**Note:** The topography points are taken from all field data, concatenated with `np.concatenate`,
and then gridded.

```python
# gridding
grav = scipy.interpolate.griddata ((data_grav[:,0],data_grav[:,1]),data_grav[:,3],(X,Y),method='nearest')
mag  = scipy.interpolate.griddata ((data_mag[:,0],data_mag[:,1]),data_mag[:,3],(X,Y),method='nearest')
sp   = scipy.interpolate.griddata ((data_sp[:,0],data_sp[:,1]),data_sp[:,3],(X,Y),method='nearest')
topo = scipy.interpolate.griddata ((np.concatenate((data_grav[:,0],data_mag[:,0])),
                                    np.concatenate((data_grav[:,1],data_mag[:,1]))),
                                   np.concatenate((data_grav[:,2],data_mag[:,2])),
                                   (X,Y),method='nearest')
```

----
## Scaling regular data to image resolution

Next, we rescale all field values to the range [0,255], the 8-bit range of a color image.
This is done in two steps:
- First, scale each field set by its min/max values to the range [1,255]. Note that we need to set the final list to unsigned integer values, using the `astype()` function with `uint8`.

```python
# scale to [1,255] range, leave 0 for NaN ...
gravmin = grav.min()
gravmax = grav.max()
grav = 254*(grav-grav.min()) / (grav.max()-grav.min())
grav += 1
grav = grav.astype('uint8')
print('min/max grav, scaled: ',gravmin,gravmax,grav.min(),grav.max())

magmin = mag.min()
magmax = mag.max()
mag = 254*(mag-mag.min()) / (mag.max()-mag.min())
mag += 1
mag = mag.astype('uint8')
print('min/max mag, scaled: ',magmin,magmax,mag.min(),mag.max())

spmin = sp.min()
spmax = sp.max()
sp = 254*(sp-sp.min()) / (sp.max()-sp.min())
sp += 1
sp = sp.astype('uint8')
print('min/max sp, scaled: ',spmin,spmax,sp.min(),sp.max())

topomin = topo.min()
topomax = topo.max()
topo = 254*(topo-topo.min()) / (topo.max()-topo.min())
topo += 1
topo = topo.astype('uint8')
print('min/max topography, scaled: ',topomin,topomax,topo.min(),topo.max())
```

- Second, we create a function `mask_data`, which returns `True` for grid locations, where original
data values are close, and `False` else. **Note:** We use a search radius to define a true box.
Then, we use the mask to set the gridded value to `0`, when there is no data point (`False`).
Thus, `0` is our **not-a-number** marker.

```python
def mask_data(x,y,data,radius):
    '''
    # mask no-data values
    '''
    nx=x.shape[0]
    ny=x.shape[1]
    mask = np.ones([nx,ny],dtype=bool)
    for i in np.arange(0,nx):
        for j in np.arange(0,ny):
            for k in np.arange(0,len(data[:,0])):
                dist = np.sqrt((x[i][j]-data[k,0])**2 + (y[i][j]-data[k,1])**2)
                if (dist < radius):
                    mask[i][j] = False
    return mask
```

```python
radius = 2*np.sqrt(dx**2 + dy**2)
print('radius: ',radius)

mask_grav = mask_data(X,Y,data_grav,radius)
grav[mask_grav] = 0

mask_mag = mask_data(X,Y,data_mag,radius)
mag[mask_mag] = 0

mask_sp = mask_data(X,Y,data_sp,radius)
sp[mask_sp] = 0
```

We plot the gridded data set.

```python
# plot data
fig,axs = plt.subplots(2,2,figsize=(8,6))
print(axs.shape)

axs[0,0].set_title('Gravity')
axs[0,0].set_ylabel('Northing [km]')
axs[0,0].set_xlim([eastmin/1000,eastmax/1000])
axs[0,0].set_ylim([northmin/1000,northmax/1000])
cbar0=axs[0,0].contourf(X/1000,Y/1000,grav)
axs[0,0].plot(data_grav[:,0]/1000,data_grav[:,1]/1000,marker='x',markersize=3,lw=0,color='black',alpha=0.5)
plt.colorbar(cbar0)

axs[1,0].set_title('Magnetics')
axs[1,0].set_xlabel('Easting [km]')
axs[1,0].set_ylabel('Northing [km]')
axs[1][0].set_xlim([eastmin/1000,eastmax/1000])
axs[1][0].set_ylim([northmin/1000,northmax/1000])
cbar1=axs[1][0].contourf(X/1000,Y/1000,mag)
axs[1][0].plot(data_mag[:,0]/1000,data_mag[:,1]/1000,marker='x',markersize=3,lw=0,color='black',alpha=0.5)
plt.colorbar(cbar1)

axs[0,1].set_title('SP')
axs[0][1].set_xlim([eastmin/1000,eastmax/1000])
axs[0][1].set_ylim([northmin/1000,northmax/1000])
cbar2=axs[0][1].contourf(X/1000,Y/1000,sp)
axs[0][1].plot(data_sp[:,0]/1000,data_sp[:,1]/1000,marker='x',markersize=3,lw=0,color='black',alpha=0.5)
plt.colorbar(cbar2)

axs[1,1].set_title('Topography')
axs[1,1].set_xlabel('Easting [km]')
axs[1][1].set_xlim([eastmin/1000,eastmax/1000])
axs[1][1].set_ylim([northmin/1000,northmax/1000])
cbar3=axs[1][1].contourf(X/1000,Y/1000,topo)
axs[1][1].plot(data_sp[:,0]/1000,data_sp[:,1]/1000,marker='x',markersize=3,lw=0,color='black',alpha=0.5)
plt.colorbar(cbar3)

plt.tight_layout()
```

## Export to image

Finally, we export the gridded data sets to the different channels:
- **red:** Gravity
- **green:** Magnetics
- **blue:** Self-potential
- **alpha:** Topography

**Note:**<br>
The regular data are stored as 2D numy array, with **rows** as first index from top to bottom,
and **columns** as second index from left to right.

However, the coordinate origin of our data set is the **lower left**.

We therefore need to flip the northing direction, as our northing coordinates generated 
with `np.linspace` start on top with the lowest values, but need to be on bottom in the image.

(In the figure above, matplotlib takes care to put the lowest northing value along the bottom).

We use the `flip()` function from the `numpy` package, flipping the first direction (`axis=0`).

```python
image_arr = np.zeros([grav.shape[0],grav.shape[1],4],dtype=np.uint8)
image_arr[:,:,0] = np.flip(grav,axis=0) #grav
image_arr[:,:,1] = np.flip(mag,axis=0) #mag
image_arr[:,:,2] = np.flip(sp,axis=0) #sp
image_arr[:,:,3] = np.flip(topo,axis=0) #topo

print (type(image_arr),image_arr.dtype,image_arr.ndim,image_arr.shape)
print(image_arr[0][0])
```

In the final image, we can see the correct location of the four data sets ...

```python
image = PIL.Image.fromarray((image_arr).astype(np.uint8))
print(image)
print (image.format,image.size,image.mode)
# show image
fig,axs = plt.subplots(1,1)
axs.imshow(image)

image.save('GP_Walpenreuth.png')
```
