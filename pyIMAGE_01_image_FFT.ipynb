# Image processing: FFT filter

In this notebook, we apply filter to images, based on the **FFT** (fast Fourier transform) methods.

As we load the image with the `Image.load` function from the `PIL` library,
we:
- first create a PIL object, `image`, 
- then a numpy array `image_arr`.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import PIL
tiny=1e-15
```

Example [follows scipy handbook](https://scipy-lectures.org/intro/scipy/auto_examples/solutions/plot_fft_image_denoise.html#sphx-glr-intro-scipy-auto-examples-solutions-plot-fft-image-denoise-py)

## Original image

Load with `Image.read` from the `PIL` package, then convert PIL object
to numpy array, using `asarray()`.

Finally plot with `imshow`.

```python
infile = 'images/moonlanding.png'
image = PIL.Image.open(infile)
print (image.format,image.size,image.mode)
# convert to ndarray
image_arr = np.asarray(image)
print (type(image_arr),image_arr.dtype,image_arr.ndim,image_arr.shape)
print(image_arr[0][0])

# show image
fig,axs = plt.subplots(1,1,figsize=(10,5))
axs.set_title('Original ('+str(image_arr.shape[0])+'x'+str(image_arr.shape[1])+')')
axs.imshow(image_arr,cmap='gray')
```

----
## FFT

Transform image to spectral domain, using the `fft2` function from the `np.fft` package.
Similar functions are in `scipy`...

The two pplots show the absolute values of the Fourier coefficients:
- **left:** In `numpy` notation, with small $k$ values, corresponding to large wavelengths,
along the edge,
- **right:** Shifted with small $k$ values (large wavelength) to centrum.

For this shift, there is a convenient function pair, `fftshift` and `ifftshift`, in the
`numpy.fft` package.

```python
# forward FFT
im_fft = np.fft.fft2(image_arr)

# plot spectrum (unshifted and shifted)
fig,axs = plt.subplots(1,2,figsize=(10,5))
axs[0].set_title('Fourier transform')
cbar0=axs[0].imshow(np.abs(im_fft), norm=LogNorm(vmin=5))
plt.colorbar(cbar0,location='bottom',shrink=0.8)
axs[1].set_title('Fourier transform (shifted)')
cbar1=axs[1].imshow(np.abs(np.fft.fftshift(im_fft)), norm=LogNorm(vmin=5))
plt.colorbar(cbar1,location='bottom',shrink=0.8)
```

----
## Filter

Create a filter in the spectral domain.

We first create wavenumber arrays for the spectral domain.

Then, we use an exponential function (similar to Gaussian filter):
$$
F = e^{-a\sqrt{k_x^2 + k_y^2}}
$$

```python
nx = image_arr.shape[1]
ny = image_arr.shape[0]
kx     = (-nx/2 + np.arange(0,nx))*2*np.pi/(2*nx)
ky     = (-ny/2 + np.arange(0,ny))*2*np.pi/(2*ny)
KX, KY = np.meshgrid(kx, ky)
```

```python
K = np.sqrt(KX**2 + KY**2)
a=2.
Fup = np.exp(-a*K)

fig,axs = plt.subplots(1,1,figsize=(6.0,4.0))
axs.set_title('Filter in spectral domain')
CS1=axs.contourf(KX,KY,Fup)
cbar = fig.colorbar(CS1,ax=axs,location="right",shrink=0.9)
```

We apply the filter in the spectral domain by multiplying it to the spectrum of the
image. 

**Careful:** As the filter is designed in the *shifted* domain, either the 
spectrum of the image or the filter have to be shifted ...

```python
# plot spectrum (unshifted and shifted)
fig,axs = plt.subplots(1,2,figsize=(10,5))
axs[0].set_title('Fourier transform')
cbar0=axs[0].imshow(np.abs(im_fft*np.fft.fftshift(Fup)), norm=LogNorm(vmin=5))
plt.colorbar(cbar0,location='bottom',shrink=0.8)
axs[1].set_title('Fourier transform (shifted)')
cbar1=axs[1].imshow(np.abs(np.fft.fftshift(im_fft)*Fup), norm=LogNorm(vmin=5))
plt.colorbar(cbar1,location='bottom',shrink=0.8)
```

----
## Reconstruct

Reconstruct the filtered image by a backward FFT.

```python
# backward FFT
image_denoised = np.fft.ifft2(im_fft*np.fft.fftshift(Fup)).real
```

```python
# show image
fig,axs = plt.subplots(1,2,figsize=(10,5))
axs[0].set_title('Original ('+str(image_arr.shape[0])+'x'+str(image_arr.shape[1])+')')
axs[0].imshow(image_arr,cmap='gray')
axs[1].set_title('Denoised (a='+str(a)+')')
axs[1].imshow(image_denoised,cmap='gray')
```
